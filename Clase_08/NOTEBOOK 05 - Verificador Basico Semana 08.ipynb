## 0. Antes de verificar: `for` y `while`

En esta semana el tema principal son los **ciclos**. Un ciclo permite repetir instrucciones sin copiarlas muchas veces.

En Python usaremos dos tipos principales:

| Ciclo | Cuando usarlo | Idea clave |
|---|---|---|
| `for` | Cuando ya tengo una coleccion de datos para recorrer. | Recorre elemento por elemento. |
| `while` | Cuando debo repetir mientras una condicion sea verdadera. | Repite hasta que la condicion cambie. |

En este caso, el ciclo principal sera `for`, porque el JSON se convierte en una lista de pacientes.


### Caracteristicas de `for`

Use `for` cuando sabe que va a recorrer una lista, un texto, un rango o cualquier coleccion.

Caracteristicas:

- Recorre una coleccion completa.
- En cada vuelta toma un elemento.
- Es ideal para listas de diccionarios.
- Evita manejar posiciones manualmente.

Estructura general:

```python
for elemento in coleccion:
    # instrucciones que se repiten
```

En esta clase:

```python
for paciente in pacientes:
    edad = paciente["edad"]
    provincia = paciente["provincia"]
```

Lectura pedagogica: **cada vuelta del ciclo representa un paciente**.


````python
# Ejemplo pequeno de for: recorrer nombres
nombres = ["Ana", "Luis", "Sofia"]

````

### Caracteristicas de `while`

Use `while` cuando no sabe exactamente cuantas veces debe repetir, pero si sabe la condicion para continuar.

Caracteristicas:

- Depende de una condicion.
- Puede repetirse cero, una o muchas veces.
- Necesita que algo dentro del ciclo cambie la condicion.
- Si la condicion nunca cambia, puede crear un ciclo infinito.

Estructura general:

```python
while condicion:
    # instrucciones que se repiten
    # algo debe cambiar la condicion
```

Ejemplos comunes:

- pedir datos hasta que el usuario escriba una opcion valida,
- repetir un menu hasta que el usuario elija salir,
- intentar leer informacion hasta que se cumpla una condicion.


````python
# Ejemplo pequeno de while: contador controlado
contador = 1

while contador <= 3:
    print("Intento", contador)
    contador += 1
````

### Cual usamos en el caso del hospital?

Usamos `for`, porque despues de leer el JSON tenemos una lista llamada `pacientes`.

```python
for paciente in pacientes:
    # analizar paciente actual
```

No usamos `while` para recorrer pacientes porque ya conocemos la coleccion que queremos procesar. Usar `while` obligaria a manejar posiciones manualmente y seria mas facil cometer errores.

Regla rapida:

- Si tiene una lista para recorrer: use `for`.
- Si necesita repetir hasta que algo ocurra: use `while`.
