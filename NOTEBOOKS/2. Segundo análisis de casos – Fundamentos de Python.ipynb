# Problema #1: Calculadora de propina inteligente

## Problema

Desarrollar un programa que solicite al usuario el total de una cuenta y el porcentaje de propina que desea dejar. El programa debe calcular el monto de la propina y el total final a pagar.

## Entrada

-El programa solicita al usuario:
El total de la cuenta y el porcentaje de propina.
(Ambos valores deben ser numéricos y pueden contener decimales)

## Proceso

1. Solicitar el total de la cuenta.
2. Solicitar el porcentaje de propina.
3. Calcular la propina utilizando la fórmula:
        Propina = Total de la cuenta * (Porcentaje/100)
4. Calcular el total final:
        Total final = Total de la cuenta + Propina
5. Se utiliza try.except para validar que los datos ingresados sean numéricos.

## Salida esperada

- El programa debe mostrar:
El monto de la propina.
El total final a pagar.

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|-----------|------------------|--------------------|--------------------|
| Caso normal | Total de la cuenta=100, Propina=10% |Propina: $10.00, Total a pagar: $110.00 |Propina: $10.00, Total a pagar: $110.00 |
| Caso límite |Total de la cuenta=$50, Propina=0% |Propina: $0.00. Total a pagar: $50.00 |Propina:$0.00, total a pagar;$50.00 |
| Caso inválido |Total de la cuenta="cien", Propina=10% |Error: Debe ingresar valores numéricos válidos. |Error: Debe ingresar valores numéricos válidos. |


```python
# Calculadora de propina inteligente
try:
    total_cuenta = float(input("Ingrese el total de la cuenta: "))
    porcentaje_propina = float(input("Ingrese el porcentaje de propina: "))
    
    propina = total_cuenta * (porcentaje_propina / 100)
    total_final = total_cuenta + propina
    
    print(f"propina: ${propina:.2f}")
    print(f"Total a pagar: ${total_final:.2f}")
    
except ValueError:
    print("Error: Debe ingresar valores numericos válidos")
    
```

# Problema #2: Detector de número par/impar

## Problema

Desarrollar un programa que solicite al usuario un número entero e indique si dicho número es par o impar.

## Entrada

-El programa solicita:
Un número entero
Se utiliza try,except para validar que el dato ingresado sea un número entero válido.

## Proceso

1. Solicitar al usuario un número entero.
2. Utilizar el operador módulo % para obtener el residuo de dividir el número entre 2.
3. Si el residuo es 0, el número es par.
4. En caso contrario, el número es impar.
5. Si el usuario ingresa in valor no nýmerico, mostrar un mensaje de error.

## Salida esperada

- El programa debe indicar si el número ingresado es par o impar.

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|-----------|------------------|--------------------|--------------------|
| Caso normal |Número=8 |El número 8 es par |El número 8 es par |
| Caso límite |Número=0 |El número 0 es par |El número 0 es par |
| Caso inválido |Número=hola |Error: Debe ingresar un numero entero valido. | Error: Debe ingresar un numero entero valido.|


```python
# Detector de número par/impar

try:
    numero = int(input("Ingrese un número entero: "))
    
    if numero % 2 == 0:
        print(f"El número {numero} es par. ")
    else:
        print(f"El número {numero} es impar. ")
        
except ValueError:
    print("Error: Debe ingresar un número entero válido")
```

# Problema #3: Clasificador de edad

## Problema

Desarrollar un programa que solicite la edad de una persona y la clasifique como menor de edad, adulto o adulto mayor según su edad.

## Entrada

- El programa solicita:
La edad de una persona (numero entero)
Se utiliza try.except para validar que el dato ingresado sea numerico.

## Proceso

1. Solicitar la edad.
2. Verificar la categoria correspondiente:
    Si la edad es menor a 18, es menor de edad.
    Si la edad está entre 18 y 64 años, es adulto.
    Si la edad es mayor o igual a 65 años, es adulto mayor.
3. Mostrar la clasificacion obtenida.
4. Si el usuario ingresa un valor no númerico, mostrar un mensaje de error.

## Salida esperada

- El program debe indicar la categoria correspondiente a la edad ingresada.

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|-----------|------------------|--------------------|--------------------|
| Caso normal |Edad=25 |La persona es adulta |La persona es adulta. |
| Caso límite |Edad=65 |La persona es adulta mayor. |La persona es adulta mayor. |
| Caso inválido |Edad=veinte |Error= Debe ingresar una edad valida. |Error= Debe ingresar una edad valida. |


```python
# Clasificador de edad

try:
    edad = int(input("Ingrese su edad: "))
    
    if edad < 18:
        print("La persona es menor de edad. ")
    elif edad < 65:
        print("La persona es adulta. ")
    else:
        print("La persona es adulta mayor. ")
        
except ValueError:
    print("Error: Debe ingresar una edad válida")
```

# Problema #4: Validador de compra con descuento

## Problema

Desarrollar un programa que solicite el nombre del cliente, el monto de la compra y si es cliente frecuente. El programa debe calcular el descuento correspondiente segun las condiciones establecidas y mostrar el total a pagar.
## Entrada

- El programa solicita:
Nombre del cliente.
Monto de la compra.
Si es cliente frecuente (sí o no)

## Proceso

1. Solicitar el nombre del cliente.
2. Solicitar el monto de la compra.
3. Solicitar si es cliente frecuente.
4. Aplicar las siguientes reglas:
    Si es cliente frecuente y la compra es mayor o igual a $100, aplicar un 10% de descuento.
    Si es cliente frecuente o la compra es mayor a %50, aplicar un 5% de descuento.
    En cualquier otro caso, no aplicar descuento.
5. Calcular el descuento y el total a pagar.
6. Mostrar resultados.


## Salida esperada

- El prgrama debe mostrar:
Nombre del cliente.
Porcentaje de descuento aplicado.
Monto descontado.
Total a pagar.

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|-----------|------------------|--------------------|--------------------|
| Caso normal |Ana, %120, si |Descuento 10%, total $108.00 |Descuento 10%, total $108.00 |
| Caso límite |Luis, $50, no |Descuento 5%, total $47.50 |Descuento 5%, total $47.50 |
| Caso inválido |Maria, cien, si |Error: Debe ingresar un monto numerico valido |Error: Debe ingresar un monto numerico valido |


```python
# Validador de compra con descuento

try:
    cliente = input("Ingrese el nombre del cliente: ")
    monto = float(input("Ingrese el monto de la compra: "))
    frecuente = input("¿Es cliente frecuente? (si/no): ").lower()
    
    if frecuente == "si" and monto >= 100:
        descuento = 0.10
    elif frecuente == "si" or monto >= 50:
        descuento = 0.05
    else:
        descuento = 0
        
    monto_descuento = monto * descuento
    total_pagar = monto - monto_descuento

    print(f"\nCliente: {cliente}")
    print(f"Descuento aplicado: {descuento*100:0f}%")
    print(f"Monto descontado: ${monto_descuento:.2f}")
    print(f"Tptal a pagar: ${total_pagar:.2f}")

except ValueError:
        print("Error: Debe ingresar un monto numerico valido.")
```
