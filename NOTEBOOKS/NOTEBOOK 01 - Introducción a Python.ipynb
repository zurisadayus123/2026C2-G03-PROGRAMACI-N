### 🔪 Slicing (Corte de Cadenas)
Acceder a partes específicas de un texto es vital. La sintaxis es `texto[inicio:fin:paso]`.

**Entendiendo los Índices:**
Recuerda que Python cuenta desde 0. También puedes usar índices negativos para contar desde el final.

| P | Y | T | H | O | N |
|---|---|---|---|---|---|
| 0 | 1 | 2 | 3 | 4 | 5 |
| -6| -5| -4| -3| -2| -1|

```python
codigo_curso = "PY-101-2026"

# Extraer el lenguaje (primeros 2 caracteres, índice 0 al 2 sin incluir el 2)
lenguaje = codigo_curso[0:2]
print(lenguaje)


# Extraer el año (últimos 4 caracteres)
anio = codigo_curso[7:]
anio = codigo_curso[-4:]

print(f"Lenguaje: {lenguaje} / Año: {anio}")


#print(f"Lenguaje: {lenguaje} | Año: {anio}")
```

### 🧼 Métodos Esenciales de Limpieza y Análisis
Las cadenas son objetos y tienen "poderes" (métodos) integrados. Para análisis de datos, estos son obligatorios:

* `.strip()`: Elimina espacios vacíos.
* `.lower()` / `.upper()`: Normaliza mayúsculas/minúsculas.
* `.split()`: Divide un texto en una lista (Crucial para CSVs).
* `in`: Verifica si un texto existe dentro de otro.

```python
texto_sucio = "   hOla MunDo  "

# 1. Limpieza y Normalización
limpio = texto_sucio.strip().lower()
print(f"Limpiado: '{limpio}'")

print(f"Limpiado: '{limpio.capitalize()}'")
```

```python
# 2. Reemplazo
fecha = "12/02/2026"
fecha = fecha.replace("2026", '2028')
print(f"Fecha Sistema: {fecha.replace('/', '-')}")
```

```python
# 3. Split (Texto a Lista) y Join (Lista a Texto)
datos_csv = "Juan,Perez,25,Costa Rica"
lista_datos = datos_csv.split(",")
print(f"Lista de datos: {lista_datos}")
```

```python
# 4. Operador 'in' (Búsqueda inteligente)
if "Costa Rica" in datos_csv:
    print("✅ El usuario es de Costa Rica")
```

### 🚀 Nivel Pro: F-Strings Avanzados
Desde Python 3.6, esta es la forma estándar. Permite formatear números (decimales, miles) directamente.

### ¿Qué aprendimos aquí?

Limpiar texto antes de comparar evita errores silenciosos. Un programa puede estar bien escrito, pero producir una decisión incorrecta si la entrada llega con espacios, mayúsculas o formatos distintos.


```python
producto = "Laptop Gamer"
precio = 1250.559
descuento = 0.10 #10%

# :.2f  -> 2 decimales
# :,    -> Separador de miles (ej: 1,000)
#Crea un mensaje que incluya el nombre del producto y su precio final con descuento aplicado, formateado con separadores de miles y dos decimales.
# Ejemplo: "Producto: Laptop Gamer | Precio Final: $1,125.50"

mensaje = f"Producto: {producto} / Precio Final ${precio * (1 - descuento):,.2f}"


print(mensaje)
```

Si el usuario escribe `50`, Python recibe `"50"`.
* ❌ `if "50" > 10`: Error o comportamiento extraño.

### 🛡️ Manejo de Errores (`try-except`)
¿Qué pasa si pedimos un número y el usuario escribe "hola"? El programa explota (crash). 
Usamos `try-except` como un **casco de seguridad**.

```python
print("--- Calculadora de Edad ---")

try:
    # Intentamos ejecutar este código peligroso
    edad_str = input("Ingresa tu edad: ")
    edad = int(edad_str)

    if edad < 0:
        print("⚠️ La edad no puede ser negativa.")
    elif edad < 18:
        print("Eres menor de edad 👶")
    else:
        print("Eres mayor de edad 💼")
        
except ValueError:
    # Si falla la conversión a int, caemos aquí
    print("❌ ¡Error! Debes ingresar un NÚMERO entero, no letras.")
```
