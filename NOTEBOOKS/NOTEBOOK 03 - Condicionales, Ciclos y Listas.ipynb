## 1. ¿Cómo funciona una lista?
Una lista es una colección **ordenada** y **mutable**. Ordenada significa que cada elemento conserva una posición. Mutable significa que podemos cambiar su contenido después de crearla.

```text
ventas = [85000, 120000, 95000, 110000]
índice      0       1      2       3
negativo   -4      -3     -2      -1
```

- `ventas[0]` obtiene el primer elemento.
- `ventas[-1]` obtiene el último.
- `ventas[1] = 125000` modifica el segundo.
- `120000 in ventas` comprueba pertenencia.
- `ventas[1:3]` crea una porción desde el índice 1 hasta antes del 3.

> Acceder a un índice inexistente produce `IndexError`. Antes de consultar una posición, debemos conocer el tamaño o comprobar que la lista no esté vacía.

````python
ventas = [85000, 120000, 95000, 110000]

print(f"Primera: {ventas[0]}")
print(f"Última: {ventas[-1]}")
print(f"Cantidad: {len(ventas)}")
print(f"Porción central: {ventas[1:3]}")

ventas[1] = 125000
print(f"Lista modificada: {ventas}")
print(f"¿Contiene 95000? {95000 in ventas}")
````

## 2. Métodos y funciones esenciales
Los **métodos** pertenecen a la lista y suelen modificarla. Las **funciones** reciben la lista y calculan un resultado.

| Herramienta | Acción | Ejemplo |
|---|---|---|
| `append(valor)` | Agrega al final | `ventas.append(75000)` |
| `insert(i, valor)` | Inserta en una posición | `ventas.insert(1, 90000)` |
| `remove(valor)` | Elimina la primera coincidencia | `ventas.remove(75000)` |
| `pop(i)` | Extrae y devuelve un elemento | `ventas.pop()` |
| `sort()` | Ordena la misma lista | `ventas.sort()` |
| `count(valor)` | Cuenta coincidencias | `ventas.count(90000)` |
| `index(valor)` | Busca el primer índice | `ventas.index(90000)` |
| `len(lista)` | Calcula la cantidad | `len(ventas)` |
| `sum`, `min`, `max` | Producen indicadores | `sum(ventas)` |

> `remove()` e `index()` producen `ValueError` si el valor no existe. Compruebe primero con `in`. `min()` y `max()` requieren una lista no vacía.

````python
ventas = [85000, 120000, 95000]
ventas.append(70000)
ventas.insert(1, 100000)
ultima_venta = ventas.pop()

if 85000 in ventas:
    ventas.remove(85000)

ventas.sort()
print(f"Ordenadas: {ventas}")
print(f"Venta extraída: {ultima_venta}")
print(f"Veces que aparece 95000: {ventas.count(95000)}")
if 95000 in ventas:
    print(f"Índice de 95000: {ventas.index(95000)}")
print(f"Total: {sum(ventas)}")
````

## 3. Ciclo `for`: recorrer una colección
Un ciclo `for` toma cada elemento de una colección, uno por uno. La variable de recorrido cambia automáticamente en cada iteración.

```python
for venta in ventas:
    print(venta)
```

Un **acumulador** conserva una suma que crece. Un **contador** aumenta cuando se cumple una condición. Ambos deben inicializarse antes del ciclo.

````python
ventas = [85000, 120000, 95000, 140000, 75000]
META_VENTA = 100000
total_ventas = 0
ventas_meta = 0
ventas_bajo_meta = 0


#COMPLETE CON EL FOR
for venta in ventas:
    total_ventas += venta
    if venta >= META_VENTA:
        ventas_meta += 1 #ventas meta + ventas meta +1
    else:
        ventas_bajo_meta += 1
print(f"Total: {total_ventas}")
print(f"Ventas que alcalzaron la meta: {ventas_meta}")

#print(f"Total: {total_ventas}")
#print(f"Ventas que alcanzaron la meta: {ventas_meta}")
````

````python
ventas = [85000, 120000, 95000, 140000, 75000]
for numero, venta in enumerate(ventas, start=1):
    print(f"Venta {numero}: ₡{venta:,.2f}")
````

## 4. Ciclo `while`: repetir con una condición

Un ciclo `while` repite un bloque **mientras una condición sea verdadera**.
Para comprenderlo, identifique siempre tres partes:

1. **Inicio:** valor antes del ciclo.
2. **Condición:** pregunta que decide si continúa.
3. **Actualización:** cambio que acerca el ciclo a su final.

Si la condición nunca cambia, el ciclo no termina.

````python
contador = 1

while contador <= 3:
    print(f"Vuelta número {contador}")
    contador += 1

print("El ciclo terminó.")
````

### Seguimiento del ejemplo

| Vuelta | Valor de `contador` | ¿`contador <= 3`? | Acción |
|---:|---:|:---:|---|
| 1 | 1 | Sí | Imprime y aumenta a 2 |
| 2 | 2 | Sí | Imprime y aumenta a 3 |
| 3 | 3 | Sí | Imprime y aumenta a 4 |
| Final | 4 | No | Sale del ciclo |

### 4.2 Repetir hasta recibir una respuesta

La condición también puede depender de una entrada. El siguiente ciclo continúa
hasta que el usuario escriba `python`. Aquí todavía no necesitamos `break` ni
`continue`: la propia condición controla la repetición.

````python
respuesta = ""

while respuesta != "python":
    respuesta = input("Escriba la palabra python: ").strip().lower()

    if respuesta != "python":
        print("Inténtelo nuevamente.")

print("Respuesta correcta.")
````
