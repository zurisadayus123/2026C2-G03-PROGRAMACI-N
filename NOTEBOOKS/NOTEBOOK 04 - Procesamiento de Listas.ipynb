## Recurso inicial: lista, tupla, diccionario o set

| Estructura | Se escribe | Cuando usarla | Ejemplo en EmprendeCR | Cuidado principal |
|---|---|---|---|---|
| Lista | `[]` | Varios datos relacionados que se recorren o conservan orden. | Ventas: `[85000, 92000, 78000]` | No depender de indices sin explicar su significado. |
| Tupla | `()` | Par o grupo fijo que no se modifica. | `("Soda San Pedro", 462000)` | No tratarla como lista mutable. |
| Diccionario | `{}` | Registro con campos con nombre. | `{"nombre": "Soda San Pedro", "ventas": [...]}` | Usar claves exactas. |
| Set | `set()` | Valores unicos, sin repetidos. | Provincias analizadas. | No usarlo si el orden importa. |

Regla rapida: lista para recorrer, diccionario para describir, tupla para pares fijos y set para quitar repetidos.
